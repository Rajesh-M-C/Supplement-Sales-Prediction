```python
# creating a copy from original dataframe for making changes
train = org_train.copy()
```

```python
#Dropping ID and #Order column from train data
train.drop(["ID","#Order"],axis=1,inplace=True)
```

```python
train.head()
```

```python
# creating a copy from original dataframe for making changes and dropping ID from test data
test = org_test.copy()
test.drop("ID",axis=1,inplace=True)
```

```python
#setting Date as index 
train.set_index("Date",inplace=True)
test.set_index("Date",inplace=True)
```

```python
test.head()
```

```python
train.head()
```

```python
#performing one hot encoding for categorical variables on train data
x = pd.get_dummies(train)
```

```python
x.head()
```

```python
# splitting dependent and independent variable
x.drop("Sales",inplace=True,axis=1)
```

```python
# Depenedent variable
y = train[["Sales"]]
```

```python
y
```

```python
test.head()
```

```python
#performing one hot encoding for categorical variables on test data
x_valid = pd.get_dummies(test)
```

```python
x_valid.head()
```

```python
# importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
#creating train test split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.70,random_state=0)
```

```python
#Model Building using XGBRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
XGB = XGBRegressor(random_state=0)
```

```python
#Fitting train data to XGBRegressor and predicting for train data and calculating the accuracy
XGB.fit(x_train,y_train)
XGB_predict = XGB.predict(x_train)
r2_score(y_train,XGB_predict)
```

```python
#Predicting for test data and calculating the score
XGB_predict_test = XGB.predict(x_test)
r2_score(y_test,XGB_predict_test)
```

```python
# predicting for validation data
predict_XGB = XGB.predict(x_valid)
```

```python
# creating a prediction data 
XGB = pd.DataFrame(predict_XGB,columns=["Sales"])
```

```python
#setting ID as index
XGB.set_index(org_test.ID,inplace=True)
```
